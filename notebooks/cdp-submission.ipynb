## 1.1: What issues is the world facing?

We are living in a truly unique time. The Covid-19 pandemic has halted industries and economies around the world, sending us into a global recession. ## 1.2: Do these issues intersect?

For most, these problems may seem completely independent and warrent individual solutions. Today we will argue the contrary.

Although not obvious, the intersection of social and climate issues has been a topic for over 40 years. The coversation began in the 1980s as a strong correlation between minority communities and toxic waste disposal sites was revealed. Through protest and reform the environmental justice movement was born with the goal of distributing environmental benefits and dangers equally among racial and income groups.

In recent years this movement has again resurfaced with more and more studies highlighting the systemic vulberability of minority groups to climate issues such as extreme heat and flooding.

However, there is hope. The number of sustainable and socially equitable investments being made is increasing rapidly. This trend is the rooted in the global rise of Environmental Social Governence (ESG) reporting and investing. ## 2.2 What are the costs?

Our next step was to understand the costs that these risks could entail. For this we turned to the coorporation questionaire which had asked companies to estimate the minimum and maximum costs that these issues could have on their company. They were also asked to specify the estamated cost to take action.

The results are shown below:

```python
# Average features of each hazard

question_2_responses = companies_climate_change_2020[
    (companies_climate_change_2020['question_number'] == 'C2.3a')
]

# pd.get_dummies(question_3_responses)
# df.groupby(["id", "year"])["value"].sum()

risk_types = [
    'Acute physical',
    'Emerging regulation',
    'Current regulation',
    'Chronic physical',
    'Market',
    'Reputation',
    'Technology'
]

question_2_responses = question_2_responses[~question_2_responses['response_value'].isin(risk_types)]

unused_columns = [
    'organization', 'survey_year', 'response_received_date', 'accounting_period_to', 
    'question_unique_reference', 'module_name', 'ors_response_id',
    'submission_date', 'page_name', 'table_columns_unique_reference', 
    'row_name', 'data_point_name', 'comments', 'data_point_id', 'question_number', 
    'column_name'
]

question_2_responses = question_2_responses.drop(unused_columns, axis=1)


question_2_responses['Account row id'] = question_2_responses['account_number'].astype(str) + question_2_responses['row_number'].astype(str)
question_2_responses
rows = []

for id in question_2_responses['Account row id'].unique():
    responses = question_2_responses[question_2_responses['Account row id'] == id]
    
    account_number = responses['account_number'].unique()[0]
    row_number = responses['row_number'].unique()[0]
    hazard_type = responses[responses['column_number'] == 3]['response_value'].head(1).item()
    max_cost = responses[responses['column_number'] == 13]['response_value'].item()
    min_cost = responses[responses['column_number'] == 12]['response_value'].item()
    action_cost = responses[responses['column_number'] == 15]['response_value'].item()
    
    rows.append([account_number, row_number, hazard_type, max_cost, min_cost, action_cost])

columns = ['Account Number', 'Row Number', 'Hazard Type', 'Max Cost', 'Min Cost', 'Action Cost']
test = pd.DataFrame(rows, columns=columns).dropna()

test['Max Cost'] = test['Max Cost'].astype(float)
test['Min Cost'] = test['Min Cost'].astype(float)
test['Action Cost'] = test['Action Cost'].astype(float)
test.head()
processed_test = test.groupby(['Hazard Type']).mean()


processed_test.head(8)

# show distribution
fig = plt.figure(figsize=(14,4))
fig.suptitle("The Cost of Climate Risks", fontsize=17)
ax1 = fig.add_subplot(111)
processed_test.head(10).drop(['Account Number', 'Row Number'], axis=1).sort_values(by=['Max Cost']).plot.bar(ax=ax1, color=[colors[4], colors[6], colors[8]])
ax1.yaxis.grid(True)
ax1.xaxis.grid(False)
ax1.set_ylabel('Averge Cost (Billion $)')
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
```
